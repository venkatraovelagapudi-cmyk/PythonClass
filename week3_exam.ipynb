```python
# the factorial finde
N=int(input())
fact=1
if N==0 or N==1:
    print(fact)
elif N<0:
    print("Enter Postive Number")
else:
    for num in range(2,N+1):
        fact=num*fact
    print(fact)
        
```

```python
N=int(input())
sum_3=[]
sum_5=[]
for num in range(1,N+1):
    if num%3==0 and num%5==0:
        continue
    elif num%3==0:
        sum_3.append(num)
    elif num%5==0:
        sum_5.append(num)
difference=sum(sum_3)-sum(sum_5)
print(abs(difference))
```
