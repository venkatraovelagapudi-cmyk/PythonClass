```python
# 1.positonal arguments
def sub(x,y):
    c=x-y
    return c
# function calling
a,b=10,30
print(sub(a,b))
print(sub(b,a))
```

```python
# positonal arguments aading of two no's
def add(x,y):
    c=x+y
    return c
# function calling
a,b=10,20
print(add(a,b))
```

```python
# positonal arguments adding of three no's
def sum_of_nos(a,b,c):
    sumis=a+b+c
    return sumis
a,b=10,20
print(sum_of_nos(a,b,20))
```

```python
# positonal arguments adding of four no's
def sum_of_nos(a,b,c,d):
    sumis=a+b+c+d
    return sumis
a,b,c,d=10,20,3,5
print(sum_of_nos(a,b,c,d))
```

```python
# 2.keyword arguments
def sub(x,y):
    c=x-y
    return c
# function calling
a,b=10,30
print(sub(x=a,y=b))
print(sub(y=b,x=a))
```

```python
# 3.default arguments
def sum_of_nos(a,b,c=0,d=0):
    sumis=a+b+c+d
    return sumis
print(sum_of_nos(10,20))
print(sum_of_nos(10,20,4))
print(sum_of_nos(20,10,20,5))
print(sum_of_nos(10,5))
```

```python
# 3.default arguments
def pro_of_nos(a=1,b=1,c=1,d=1):
    sumis=a*b*c*d
    return sumis
print(pro_of_nos(5,5))
print(pro_of_nos(12,2,4))
print(pro_of_nos(5,3,2,1))
print(pro_of_nos(10))
```

```python
#5.variable length keyword argument by default this treated as dictionary
def details(name,**kwargs):
    info={}
    info['Name']=name
    info['Class']=kwargs.get("cls","Noclass")
    if 'cls' in kwargs:
        kwargs.pop('cls')
    marks=kwargs.values()
    info['avg']=0
    if marks:
        avg=sum(marks)/len(marks)
        info['avg']=avg
    return info

print(details("Srinu"))
print(details("Srinu",cls=10))
print(details("Srinu",cls=10,m1=50,m2=89,m3=69))
print(details("Srinu",cls=10,m1=90,m2=67,m3=80,m4=23,m5=98))
```

```python
def student_avg(name, cls, **kwargs):
    return {
        "name": name,
        "class": cls,
        "average": sum(kwargs.values()) / len(kwargs)
    }
print(student_avg("Venkat", "Btech", math=10, phys=20, bio=20))
```
