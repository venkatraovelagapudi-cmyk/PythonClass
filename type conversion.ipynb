```python
 #type conversion nd type costing @ input formating
#int-string,float,bool
num=int(input())
print(type(num))
val=float(num)
print(type(val),val)
string_num=str(num)
print(type(string_num),string_num)
print(bool(num))
```

```python
#float-int,string,bool
val=float(input())
integer_value=int(val)
print(type(integer_value),integer_value)
print(str(val)[:4])
print(bool(val))
```

```python
#list
a = list(input())
print(type(a), a)

t = tuple(a)
print(type(t), t)

s = set(a)
print(type(s), s)

print(bool(a))
```

```python
#Tuple
t = tuple(input())
print(type(t), t)

l = list(t)
print(type(l), l)

s = set(t)
print(type(s), s)

print(bool(t))
```

```python
#set
s = set(input())
print(type(s), s)

l = list(s)
print(type(l), l)

t = tuple(s)
print(type(t), t)

print(bool(s))
```

```python
#Dictionary
d = dict([("a", 1), ("b", 2)])
print(type(d), d)

keys = list(d.keys())
print(type(keys), keys)

values = tuple(d.values())
print(type(values), values)

print(bool(d))
```
