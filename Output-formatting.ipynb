```python
#using old style formatting [using commas]
age=20
name="babu"
print("my name is",name,"and i am a",age,"years old")
```

```python
# using module operator
#%s for string ,%d for integer ,%f for float
print("my name is %s and i am a age %d years old" %(name,age))
```

```python
# using dot format method
print("my name is {} and i am a {} years old".format(name,age))
print("my name is {} and i am a {} years old".format(age,name))
```

```python
# using F -Format method
age=20.66541654
name ="venkatrao"
print(f"my name is {name} and i am a {age} years old")
print(f"my name is {name} and i am a {age:.2f} years old")
```
