### list methods
- append(elem):it add the element at the end of the list but only a single value.
-extend(list_ele): it adds all list elements at the end of the list but it adds multiple values.
-insert(pos,val)it inserts the value at the particular postion.
# removing elements from list 
-remove(val): it remves the first occurance vaule if value is present in lsit,otherwise value error
-pop():it remove and returns the last element in the list
-pop(position): it removes and returns specified position element from the list.
# index finding for an element
index(val): it return the first occurance value index position of a value
# sorting list
sort(): it sorts the original list into ascending order
sort(reverse=True): it sorts the original list into descending order
# counting

```python
#adding elements to the list
lst=[1,2,3,4,2,2]
lst.append(1564)
print(lst)
lst.append([54,56,89,486])
print(lst)
lst.extend([54,56,89,486])
print(lst)
lst.insert(1,564)
print(lst)
print(lst.count(2))
```

```python
#removing elements from lsit
lst=[10,20,2,3,4.85]
lst.remove(2)
print(lst)
#lst.remove(100) # it raise vauerror because value not in the lsit
val=lst.pop()
print(val,lst)
lst.clear()# it removes all elements from the list and lsit is empty.
print(lst)
del lst
print(lst)
```
