```python
#local scope and global scope
a=10
def func(x):
    z=x+a
    return z
print(func(a))
print(a)
```

```python
#local scope and global scope
a=10
def func(x):
    global a
    a=x+a
    return a
print(func(a))
print(a)
```

```python
# global varibales using in local and non-local scope
x=10# global
def outer(a):#local
    c=a+x
    def inner(b): # non local
        k=b+x
        print("this is from inner function",k)
    inner(c)
    print("this is from outer function",c)
outer(x)
print(x)
```

```python
# global varibales using in local and non-local scope but we are updating x value in local scope 
#to update it raisies an error to avoid that we global keyword
x=10# global
def outer(a):#local
    global x
    x=a+x
    c=x+2
    def inner(b): # non local
        c=c+b
        print("this is from inner function",c)
    inner(x)
    print("this is from outer function",c)
outer(x)
print(x)
```

```python
# global varibales using in local and non-local scope but we are updating x value in local scope 
#to update it raisies an error to avoid that we global keyword
x=10# global
def outer(a):#local
    global x
    x=a+x
    c=x+2
    def inner(b): # non local
        nonlocal c
        c=c+b
        print("this is from inner function",c)
    inner(c)
    print("this is from outer function",c)
outer(x)
print(x)
```

```python
#pass by reference (all mutable datatypes are by the reference) it shares the meory location -> original value changes
lst=[1,2,3,4]
def add_ele(arr):
    arr.append(100)
    return arr
print(add_ele(lst))
print(lst)
```

```python
#pass by value (all immutable datatypes are by the value) it shares the value -> original values does not change
s="Codegnan"
def replace(string,old,new='@'):
    res=string.replace(old,new)
    return res
print(replace(s,'n','Z'))
print(s)
```

```python
#pass by value : in pass by valuem, the arguments objects are sare the object (argument,variable) copy value to function defination parameter.
# -> all mutable datatypes(ie string,tuple,int) behave like pass by value 
# -> after function execution the original value not effets

#pass by reference: in pass by reference to function defination parameter,so after the function execution original values will effects.
#-> all mutable datatypes(i.e list,set,dictionary) behave like pass by reference
```
