```python
#ord()-> it returns ascii value of a character
#chr()-> it returns charcter for specified ascii value
print(ord('A'))
print(chr(65))
```

```python
n=4
for i in range(n):
    for j in range(n):
        if i==j or i>=j:
            print(chr(65+j),end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
#len(),min(),max(),sum()
#len= returns total number of elements in sequence
#min()=it returns minimum value from the sequence
#max()=it returns maximum value from the sequence
#add() = it add the all elements when sequence contains numeric elements

lst=[1,2,3]
lst2=['hi','hello','how','are','you']
print(len(lst),min(lst),max(lst),sum(lst))
print(len(lst2),min(lst2),max(lst2))
```

```python
#sorted() this fucntion returns the sequence either acseding and descending order
lst=[1,2,3,4]
lst2=['hi','hello','how','are','you']
res=sorted(lst,reverse=True) #sorted creates new list but # sort method in list modifies old list
print(res)
print(lst)
```

```python
# enumerate(): it returns the indexx values along with sequence elements
lst2=['hi','hello','how','are','you']
res=list(enumerate(lst2))
print(res)
```

```python
#round -it rounds the floating values
print(round(10.2357,2))
print(f"{10.2357:.2f}")
```

```python
#filter(func,seq): it filter the sequence
def even(x):
    if x%2==0:
        return x
    return None
lst=[1,2,3,4,5,6]
print(list(filter(even,lst)))
```

```python
#map(func,seq) -it applies functionality  on each eelement in sequence 
def add(x,y):
    return x+y
print(list(map(add,[1,2,3],[1,2,3])))
```

```python
#zip() it returns multiple sequence as single sequence
lst1=[1,2,3]
lst2=['hi','hello','how','are','you']
print(list(zip(lst1,lst2)))
```

```python
#zip() it returns multiple sequence as single sequence
lst1=[1,2,3]
lst2=['hi','hello','how','are','you']
print(list(zip(lst1,lst2,[10,20])))
```
