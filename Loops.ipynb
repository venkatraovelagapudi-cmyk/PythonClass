```python
#while loop for even and odd
num=int(input())
num1=int(input())
even=[]
odd=[]
while num<=num1:
    if num%2==0:
        even.append(num)
        num+=1
    else:
        odd.append(num)
        num+=1
print(even)
print(odd)
```

#### loop control statements 
1.continue
2.break

```python
#continue
for i in range(1,5):
    continue
    print(i)
else:
    print("program done")
```

```python
#continue
for i in range(1,5):
    if i==3:
        continue
    print(i,end=" ")
else:
    print("program done")
```

```python
#break
for i in range(1,5):
    if i==3:
        break
    print(i,end=" ")
else:
    print("program done")
```

```python
#check weather target elemnt is exsits in list of elements or not
target_ele=int(input())
arr=list(map(int,input().split()))
for i in arr:
    if i==target_ele:
        print(f"{target_ele} exsits in {arr}")
        break
else:
    print(f"{target_ele} not exsits in {arr}")
```

```python
#check weather target elemnt is exsits in list of elements or not using indexing method
target_ele=int(input())
arr=list(map(int,input().split()))
for num in range(len(arr)):
    if arr[num]==target_ele:
        print(f"{target_ele} exsits in {arr}")
        break
else:
    print(f"{target_ele} not exsits in {arr}")
```

```python
#check weather target elemnt is exsits in list of elements nd using indexing and printing index number
target_ele=int(input())
arr=list(map(int,input().split()))
for num in range(len(arr)):
    if arr[num]==target_ele:
        print(f"{target_ele} exsits in index {arr[num]}")
        break
else:
    #print(f"{target_ele} not exsits in {arr[num]}")
    print("-1")
```

```python
num=int(input())
res=1
if num==0 or num==1:
    print(f'factorial of {num} is {res}')
else:
    for i in range(2,num+1):
        res=res*i
    print(f'factorial of {num} is {res}')
```
