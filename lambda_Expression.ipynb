```python
#Anonymous Function
#lambda
#find the square of a number
square=lambda x:x*x
print(square(2))
print(square(4))
print(type(square))
```

```python
i=3
even_odd= lambda x:x%2==0
print(even_odd(5))
print(even_odd(6))
```

```python
i=2
even_or_odd=lambda x:"true" if i%2==0 else "false"
print(even_or_odd(11))
```

```python

square=lambda x:x**0.5 if x%2==0 else x**2
print(square(11))
print(square(10))
```

```python
#check wheather the number is perfect square or not
perfect_sq=lambda x:"Perfect Square" if x**0.5==int(x**0.5) else "Not Perfect"
print(perfect_sq(10))
print(perfect_sq(16))
```

```python
# filter the even numbers from list using filter and lambda function
lst=[1,2,3,4,5,6,9,20]
print(list(filter(lambda x:x%2==0,lst)))
```

```python
# filter the square of ecah number in the lsit using map and lambda function 
lst=[1,2,3,4,5,6,9,20]
print(list(map(lambda x:x**2,lst))) #map(func,seq)
```
