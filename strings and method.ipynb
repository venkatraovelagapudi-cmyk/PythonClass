### String Methods
-Case Conversion Methods
    -upper():it converts the entire string to uppper case letters
    -lower():it converts the entire string to lower case letters
    -title():it converts the every first characater in each word into uppercase and remaining all are into lower case
    -swapcase():it converts the lower case letters to upper case and vice-versa
    -capitalize():it converts the first character of string into uppercase abd remaninig all are into lower case

```python
s="The sky is blue and Ocean is BLUE"
print(s.upper())
print(s.lower())
print(s.title())
print(s.swapcase())
print(s.capitalize())
```

```python
#substring index finding
#find(): it return the index position of the first occuring substring
s="Codegnan"
print(s.find('n'))
print(s.find('gn'))
print(s.find('Code'))
print(s.find('N')) #if substring does not found in string it returns -1 by defaultly in find()
print('-'*20)
print(s.index('g'))
print(s.index('a'))
print(s.index('N'))#if substring does not found in string it returns an error called 'valueerroe" in index()
```

- **White space or pattern removing from string***
    - strip(): it removes white spaces from the left and right of the string
    -lstrip():it removes white spaces only from the left of the string
    -rstrip(): it removes white spaces only from the right of the string

```python
s="    Codegnan     "
st=("@@@@ #@   Code gnan @# @")
print(s.strip())
print(s.lstrip())
print(s.rstrip())
print(st.strip('@'))
print(st.lstrip('@'))
print(st.rstrip('@'))
```

```python
s="The sky is blue and Ocean is blue"
print(s.replace('is','ISs'))
x="ABCBCBC"
print(x.replace('CBC','@'))
print(s.split())
print(s.split("blue")) # it removes that word and spilt the words in that particualr position
```

#### String Checking Methods
-This all ethods return boolean values as result
    -islower(); it returns true if entire string is lower
    -isupper():it returns true if entire string is uppercase
    -istitle():it returns true if entire string is title manner
    -isalpha():it returns true if entire string is contains alphabets 
    -isalnum():it returns true if entire string contains alphabets and digits
    -isdigit():it returns true only if entire string contains digits
    -startswith():it returns true if string startswith sub_string
    -endswith():it returns true if string endsswith sub_string
    

```python
print("code".islower())
print("CODE".isupper())
print("Code".istitle())
print("code123".isalnum())
print("code".isalpha())
print("123".isdigit())
print("codegnan".startswith('c'))
print("codegnan".endswith('n'))
```
