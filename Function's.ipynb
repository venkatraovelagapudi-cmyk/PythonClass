```python
# check wheather number is even of odd using functions
#def even_odd(n:int)->str:
def even_odd(n):
    if n%2==0:
        return "even"
    else:
        return "odd"
print(even_odd(5))
print(even_odd(12))
```

```python
# return the sum all elements in list 
def sum_ele(lst:list):
    return sum(lst)
print(sum_ele([1,2,3,4,5,6]))
print(sum_ele([4,5,6,7]))
```

```python
# find length of the list
def len_ele(lst:list):
    return len(lst)
print(len_ele([1,2,3,4,5,6]))
print(len_ele([4,5,6,7]))
```

```python
# find the total even numbers in list
def lst(n):
    even_lst=[]
    for num in range(1,n+1):
        if num%2==0:
            even_lst.append(num)
    return len(even_lst)
print(lst(10))
print(lst(25))
```

```python
#find total odd numbers in list 
def lst(n):
    even_lst=[]
    for num in range(1,n+1):
        if num%2!=0:
            even_lst.append(num)
    return len(even_lst)
print(lst(10))
print(lst(25))
```

```python
#find the total vowels count in string
def string(n):
  count=0
    for st in n:
        if st in "aeiouAEIOU":
            count+=1
    return count
print(string("venkat"))
print(string("srinubabusir"))
```

```python
#find the difference between consonants and vowels count from a string
def string(n):
    vowels_count=0
    consnant_count=0
    for st in n:
        if st in "aeiouAEIOU":
            vowels_count+=1
        else:
            consnant_count+=1
    difference=consnant_count-vowels_count
    return difference
print(string("venkat"))
print(string("srinubabusirfghj"))
print(string("pythonclass"))
```

```python
# find count of special character in a string 
def string(n):
    special_count=0
    for st in n:
        if not st.isalnum():
            special_count+=1
    return special_count
print(string("venk^&*(at"))
print(string("srinubabusirfgh@j"))
print(string("pythonclass"))
```
