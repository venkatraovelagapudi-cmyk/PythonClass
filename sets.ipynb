### set MEthods
- add(val): it adds the value to set of the value is already not exsits in set
- update(new_set):it adds all new set elements
- remove(val):it removes the specified value from the set if it is present in set, otherwise"Keyerror"
- discard(val): it remove the value if it present in set, otherwise nothing
- union(new_set):it returns the all unquie elements from sets.
- intersection(new_set):it returns the only common elements from all sets
- defference(new_set): it returns the set elements those are not in another set. (a-b) or (b-a)
- symmetric_deffrence(new_set): it returns the common elements from the sets (returns all non common elements)

```python
# add() 
se={1,2,3,4,5,6}
se.add(10)
print(se)
se.add(5)
print(se)
#update
se.update({1,2,300,500,10,125,1658,4535,463})
print(se)
```

```python
#remove(val),discard(val)
se={1,2,3,4,5,6}
se.remove(2)
print(se)
#se.remove(2) #KeyError:
```

```python
#discard
se={1,2,3,4,5,6}
se.discard(6)
print(se)
se.discard(6)  # doesnt rasies any error
print(se)
```
