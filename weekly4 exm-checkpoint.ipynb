```python
# 13
def print_pattern(n):
    for i in range(n):
        for j in range(n):
            if i>=j:
                print(i+1,end=" ")
        print()
print_pattern(5)
```

```python
# 15
def pattern_program(n):
    for i in range(n):
        for j in range(n):
            if i%2==0:
                print("+",end=" ")
            else:
                print("*",end=" ")
        print()
pattern_program(5)
```

```python
# 11
def calculate_discount(price,discount=10):
    result=price-price//discount
    return result
print(calculate_discount(500,10))
print(calculate_discount(800,20))
```

```python
#12 
def find_largest(*arguments):
    max_element=arguments[0]
    for num in arguments:
        if num > max_element:
            max_element=num
    return max_element
print(find_largest(2,30,15,9))
print(find_largest(12,300,5,9))
print(find_largest(12,300,5,900))
```
