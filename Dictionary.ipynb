#### dictionary methods
-keys(): it returns list dictionary keys
-values(): it returns list dictionary values
-items(): it returns list of tuples of key-value pairs 
-get(key,default_val): it returns the values of key if it is present in dictionary,otherwise default value
-update (new_dictinary): dictionary updates with new_dictionary items
-pop(key):it returnsn the value of key and removes the item from dictionary
-popitem():it return and remove the last item from dictionary

```python
details={'name':'srinu','course':'python','age':20,'name':'babu'}
keys=details.keys()
values=details.values()
items=details.items()
print(type(keys))
print(keys)
print("\n")
print(type(values))
print(values)
print("\n")
print(type(items))
print(items)
print("\n")
print(list(keys))
```

```python
details={'name':'srinu','course':'python','age':20,'name':'babu'}
print(details.get('name'))
print(details.get('age',50))
print(details.get('location')) #None
print(details.get('location','location key not present in dictionary'))
```

```python
#updating of the dictionary with an another dictionary
student={'name':'venkat','age':25,'course':'PFS','room-no':5}
new_dict={'age':50,'institute':'codegnan','location':'Hyderabad'}
student.update(new_dict)
print(student)
```

```python
#pop(key),popitem()
student={'name':'venkat','age':25,'course':'PFS','room-no':5}
print(student.pop('name'))
print(student.popitem())
```
